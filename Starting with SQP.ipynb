## Single parameter Equation

```python
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
                                             # f=((x^2)-4)^2
myzero = np.zeros(40)                        # creating an array of 40 zeros
t = np.linspace(-4, 4, 40)                   # creating array within a range that includes solution also
fun = lambda x: ((x[0])**2 - 4)**2           # x[0] is an independent var.
                                             # Now f=((x^2)-4)^2, f can be any function so that y=f
res = minimize(fun, (-0.01), method='SLSQP') #(n) in the minimize argument is the initial guess of solution
                                             # same as how we did during iterative sol(analytical) on paper
                                             
print("Result :\n",res)

plt.plot(t,((t)**2 - 4)**2)                  # Plotting t vs f(t)
plt.plot(t,myzero)                           # Plotting a straight line
plt.show()       
```

## 3-D representation of the double parameter equation.

```python
x = np.outer(np.linspace(-4, 4, 40), np.ones(40))
y = x.copy().T
z=((x)**2 - 4)**2 + (y - 9)**2
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
                                        #Plotted for  x vs y vs z(x,y)            
ax.plot_surface(x, y, z)
plt.show()
```

## New equation - 2 parameters

## 3D plot for the same equation

```python
x = np.outer(np.linspace(-4, 4, 40), np.ones(40))
y = x.copy().T # transpose
#z = (np.sin(x **2) + np.cos(y **2) )
z=100*(y-x*x)**2 + (1.0-x)**2
 
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
ax.plot_surface(x, y, z) 
# show plot
plt.show()
```

```python
from scipy.optimize import minimize
fun = lambda x: (x[0] - 1)**2 + (x[1] - 2.5)**2  #x and y
cons = ({'type': 'ineq', 'fun': lambda x:  x[0] - 2 * x[1] + 2},
        {'type': 'ineq', 'fun': lambda x: -x[0] - 2 * x[1] + 6},
        {'type': 'ineq', 'fun': lambda x: -x[0] + 2 * x[1] + 2})
res = minimize(fun, (2, 0), method='SLSQP',constraints=cons)
res
```
